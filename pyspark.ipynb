```python
#create new data set and use joins

left = pd.DataFrame({'CustomerID': [1, 2, 3],
    'Name': ['Akshay', 'yash', 'prerana'],
    'Amount': [100, 200, 150]})
```

```python
right = pd.DataFrame({ 'CustomerID': [2, 3, 4],
    'Email': ['sainiakshay823@gmail.com', 'yash@gmail.com', 'prerana@gmail.com'],
    'Country': ['USA', 'UK', 'Canada']})
```

```python
lefts = spark.createDataFrame(left)
rights = spark.createDataFrame(right)
```

```python
lefts.show()
rights.show()
```

join is a fundamental operation used to combine two DataFrames based on a common column or condition

```python
#use left join
lefts.join(rights,on="CustomerID", how='left' ).orderBy('CustomerID',ascending=True).show()
```

Returns all rows from the right DataFrame and matched rows from the left DataFrame

```python
lefts.join(rights,on="CustomerID", how='right' ).orderBy('CustomerID',ascending=True).show()
```

Returns only the rows where there is a match in both DataFrames.

```python
lefts.join(rights,on="CustomerID", how='inner' ).orderBy('CustomerID',ascending=True).show()
```

Returns all rows when there is a match in one of the DataFrames

```python
lefts.join(rights,on="CustomerID", how='full' ).orderBy('CustomerID',ascending=True).show()
```
